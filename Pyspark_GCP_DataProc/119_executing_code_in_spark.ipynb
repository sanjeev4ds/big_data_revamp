```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
.appName("119_Spark_Sessions") \
.master("yarn") \
.getOrCreate()
```

```python
spark
```

```python
data = ["word1 word2 word3",
        "word2 word3 word2",
        "word1 word3 word1",
        "word3 word2 word3"]
```

```python
rdd = spark.sparkContext.parallelize(data)
```

```python
rdd.collect()
```

```python
hdfs_file_path = "/tmp/inputhdfsdbz.txt"
```

```python
rdd_input_file = spark.sparkContext.textFile(hdfs_file_path)
```

```python
rdd_input_file.collect()
```

```python
rdd_input_file = rdd_input_file.map( lambda x: x.split(" "))
rdd_input_file.collect()
```

```python
filtered_rdd = rdd_input_file.filter(lambda x: "word1" in x )
filtered_rdd.collect()
```
